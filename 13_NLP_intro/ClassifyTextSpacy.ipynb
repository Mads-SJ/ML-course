```python
import spacy
```

```python
# Load the installed model "en_core_web_sm"
nlp = spacy.load("en_core_web_md")
```

```python
# So what is this text about?
```

```python
doc2 = nlp("John McCain and I visited the Apple Store in Manhattan.")
```

```python
for item in doc2.ents:
    print(item)
```

```python
for item in doc2.ents:
    print (item.text, item.label_)
```

```python
doc3 = nlp(open("EightThings.txt").read())
```

```python
for item in doc3.ents:
    print (item.text, item.label_)
```

```python
# Again, Dont look into the text by opening the text in an editor :) 
# Only use Spacy code to look 
# What is this text about ?
# Google what you can do, and ask chatGPT.
# So here you go...

print("\nNoun Phrases:")
for chunk in doc3.noun_chunks:
    print(chunk.text)
```

```python
print("\nVerbs:")
for token in doc3:
    if token.pos_ == "VERB":
        print(token.text)
```

```python
for sents in doc3.sents:
    print(sents)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

filtered_words = [token.text.lower() for token in doc3 if not token.is_stop and token.pos_ == "NOUN"]
print(filtered_words)
wordcloud = WordCloud().generate(' '.join(filtered_words))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
